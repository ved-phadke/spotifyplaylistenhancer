# Creating and Training Our Model
Here we actually create our model and train it by fitting it to the training data.

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (20,11)),
    keras.layers.Dense(200, activation = tf.nn.relu),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(200, activation = tf.nn.relu),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(11, activation = tf.nn.tanh)
])
```

Loss Functions: https://www.tensorflow.org/api_docs/python/tf/keras/losses

Optimizers: https://www.tensorflow.org/api_docs/python/tf/keras/optimizers



```python
model.compile(
    optimizer = 'adam',
    loss = 'MeanSquaredError',
    metrics = ['CosineSimilarity']
)
```

# Song Retrieval:

The model is cool and all, but we also want to be able to actually recommend a song. Once we have the predicted features, we will look in our million song playlist to find the song with the closest features. These cells import our possible output songs and normalize them to fit with our

```python
million_song_df = pd.read_csv('/content/drive/MyDrive/Spotify Playlist Enhancer/Deep Learning/tracks_features.csv')
million_song_df.head()
```

```python
million_song_df.columns
```

We use a modified sigmoid to scale our data. Our data and predicted song are all normalized so we need to normalize this set.

```python
def sigmoid(x):
  return (2 / (1 + np.exp(-x))) - 1
```

```python
song_features = million_song_df[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',]].to_numpy()
audio_features = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 
                  'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',]
n = len(song_features)
song_features_array = [[] for i in range(n)]
for i in range(n):
  #subtract 95 because 95 is average tempo of a song
  song_features[i][10] = song_features[i][10] - 95
  song_features_array[i] = [sigmoid(feature) for feature in song_features[i]]
song_features_array = np.array(song_features_array)
```

```python
song_features_array.shape
```

```python
song_features_array[941]
```

```python
def get_song(desired_song, features, song_df):
  n = len(features)
  most_similar_song = [100000,0]
  #similarity, position
  for i in range(n):
    #similarity = np.dot(features[i], desired_song)/(np.linalg.norm(desired_song)*np.linalg.norm(features[i]))
    similarity = np.linalg.norm(features[i] - desired_song)
    if similarity < most_similar_song[0]:
      most_similar_song[0] = similarity
      most_similar_song[1] = i
  print(f'Found a song with these features: {features[most_similar_song[1]]}')
  found_song = song_df.loc[most_similar_song[1]]
  return found_song
```

```python
def get_recommendation(playlist_link, model):
  playlist_URI = playlist_link.split("/")[-1].split("?")[0]
  track_uris = [x["track"]["uri"] for x in sp.playlist_tracks(playlist_URI)["items"]]
  selected_tracks = random.sample(track_uris, k = 20)
  audio_features = [[] for j in range(20)] 
  for i in range(20):
    song_id = selected_tracks[i].split(':')[2]
    song_characteristics = list(sp.audio_features(song_id)[0].values())[0:11]
      
        #the first 10 song features. 10th is tempo, this might not be needed
    song_characteristics[10] = song_characteristics[10] - 95
        #subtracting tempo by the average song tempo so that more information shows up after scaling
    song_characteristics = list(map(sigmoid, song_characteristics))
        #normalizing the data
    audio_features[i] =  song_characteristics

    
  predicted_features = model.predict(np.array(audio_features))
  print(f'Predicted Features: {predicted_features[0]}')
  found_song = get_song(predicted_features[0],song_features_array, million_song_df)
  print( f'Our model recommends the song {found_song[1]} by {found_song[4]}. It can be found at')
  return f'https://open.spotify.com/album/{found_song[3]}?highlight=spotify:track:{found_song[0]}'
```

# Hyper Parameter Tuning:

```python
import tensorflow as tf
from tensorboard.plugins.hparams import api as hp
```

```python
HP_NUM_UNITS = hp.HParam('num_units', hp.Discrete([1, 2, 4, 8, 10, 20, 30]))
HP_DROPOUT = hp.HParam('dropout', hp.RealInterval(0.1, 0.2))
HP_OPTIMIZER = hp.HParam('optimizer', hp.Discrete(['adam', 'sgd']))

METRIC_ACCURACY = 'accuracy'

with tf.summary.create_file_writer('logs/hparam_tuning').as_default():
  hp.hparams_config(
    hparams=[HP_NUM_UNITS, HP_DROPOUT, HP_OPTIMIZER],
    metrics=[hp.Metric(METRIC_ACCURACY, display_name='Accuracy')],
  )
```

```python
def train_test_model(hparams):
  model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(hparams[HP_NUM_UNITS], activation=tf.nn.relu),
    tf.keras.layers.Dropout(hparams[HP_DROPOUT]),
    tf.keras.layers.Dense(11, activation=tf.nn.tanh),
  ])
  model.compile(
      optimizer=hparams[HP_OPTIMIZER],
      loss='MeanSquaredError',
      metrics=['CosineSimilarity'],
  )

  model.fit(training_tracks, training_hidden, epochs=1) # Run with 1 epoch to speed things up for demo purposes
  _, accuracy = model.evaluate(testing_tracks, testing_hidden)
  return accuracy
```

```python
def run(run_dir, hparams):
  with tf.summary.create_file_writer(run_dir).as_default():
    hp.hparams(hparams)  # record the values used in this trial
    accuracy = train_test_model(hparams)
    tf.summary.scalar(METRIC_ACCURACY, accuracy, step=1)
```

```python
session_num = 0

for num_units in HP_NUM_UNITS.domain.values:
  for dropout_rate in (HP_DROPOUT.domain.min_value, HP_DROPOUT.domain.max_value):
    for optimizer in HP_OPTIMIZER.domain.values:
      hparams = {
          HP_NUM_UNITS: num_units,
          HP_DROPOUT: dropout_rate,
          HP_OPTIMIZER: optimizer,
      }
      run_name = "run-%d" % session_num
      print('--- Starting trial: %s' % run_name)
      print({h.name: hparams[h] for h in hparams})
      run('logs/hparam_tuning/' + run_name, hparams)
      session_num += 1
```
