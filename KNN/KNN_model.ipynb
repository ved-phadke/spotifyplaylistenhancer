#General Process for Spotify Recommendation (Classical Approach)

1. Subset dataset into necesary columns/features
2. Normalize features to ensure that distance calculation is correct
3. Given a playlist ID, retrieve the songs in the playlist and compute the average feature values to represent the playlist as a vector
4. Represent a playlist as a vector in the same format as songs in the reference dataset
5. Use KNN with some distance metric to recommend the K closest songs

```python
spotify_df = pd.read_csv("tracks_features.csv")
spotify_df.columns
```

```python
len(spotify_df)
```

```python
features = spotify_df[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']].dropna().to_numpy()
len(features)
```

```python
scaler = StandardScaler()
scaler.fit(features)
features_scaled = scaler.transform(features)
```

```python
knn_model = NearestNeighbors(n_neighbors=10)
knn_model.fit(features_scaled)
```
